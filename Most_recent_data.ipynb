```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("dane/gini/gini_0.csv", delimiter = ",", on_bad_lines='skip')
```

```python
df.info
```

```python
df["OBS_VALUE"].head()
```

```python
df = df[["REF_AREA_LABEL", 'TIME_PERIOD', "OBS_VALUE"]]
df
```

```python
df_sorted = df.sort_values(by=['REF_AREA_LABEL', 'TIME_PERIOD'], ascending=[True, False])
df
```

```python
df_latest = df_sorted.drop_duplicates(subset=['REF_AREA_LABEL'], keep='first')
df_latest
```

```python
df = df_latest[['REF_AREA_LABEL', 'OBS_VALUE']]
df
```

```python
df = df.rename(columns={
    'REF_AREA_LABEL': 'Country Name',
    'OBS_VALUE': 'Gini'
})
```

```python
df.info()
```

```python
znalezione_wartosci = []

for index, row in df.iterrows():
    znaleziona = None
    
    for rok in range(2024, 2000, -1):
        col_name = f"{rok}"
        if col_name in df.columns and row[col_name] != "--":
            znaleziona = row[col_name]

    znalezione_wartosci.append(znaleziona)

df["GDP"] = znalezione_wartosci

print(df[["Country Name", "GDP"]])
```
